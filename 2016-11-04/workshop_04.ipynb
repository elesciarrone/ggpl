```python
from pyplasm import *
from fpformat import *

def draw_roof():

    #lista dei vertici
    vertexs = [[0, 0, 0], [4, 0, 0], [2, 1.5, 1.25], [2, 1.5, 0], [0, 10, 0], [4, 10, 0], [2, 8.5, 1.25], [2, 8.5, 0],
               [4, 7, 0], [7, 7, 0], [6, 5, 1], [7, 3, 0], [4, 3, 0], [2.345, 5, 1], [2.345, 5, 0], [6, 5, 0]]

    #lista delle celle
    cells = [[1, 2, 3, 4], [5, 6, 7, 8], [1, 3, 4, 5, 7, 8], [2, 4, 13, 15], [6, 8, 9, 15], [4, 8, 15], [10, 11, 12, 16],
             [11, 12, 13, 14, 15, 16], [9, 10, 11, 14, 15, 16]]

    roof = MKPOL([vertexs, cells, None])

    return roof


def ggpl_roof_builder(hpc):

    #definizione dello scheletro della struttura
    skeleton1 = OFFSET([.1, .1, .1])(SKEL_1(hpc))
    skeleton2 = UKPOL(SKEL_2(hpc))

    vertexs = skeleton2[0]
    cells = skeleton2[1]


    cells2vertexs = []
    for c in cells:
        list = []
        for i in c:
            list.append(vertexs[i - 1])
        cells2vertexs.append(list)

    app = []
    for cvc in cells2vertexs:
        list = []
        for c in cvc:
            l = []
            for i in c:
                x = float(fix(i, 3))
                l.append(x)
            list.append(l)
        app.append(list)

    cells2vertexsClean = []
    for cv in app:
        for i in cv:
            if i not in cells2vertexsClean:
                cells2vertexsClean.append(i)
            else:
                i = cells2vertexsClean.index(i)

    cellsClean = []
    for cv in app:
        list = []
        for i in cv:
            list.append(cells2vertexsClean.index(i) + 1)
        cellsClean.append(list)

    cells2vertexsFinal = []
    trovato = 0
    for cvc in cells2vertexsClean:
        if (cvc[2] > 0):
            cells2vertexsFinal.append(cvc)
        else:
            for cvc2 in cells2vertexsClean:
                if cvc2[2] > 0:
                    if cvc2[0] == cvc[0] and cvc2[1] == cvc[1]:
                        cells2vertexsFinal.append(cvc2)
                        trovato = 1
            if not trovato:
                cells2vertexsFinal.append(cvc)
            trovato = 0

    #definizione del tetto
    structure = MKPOL([cells2vertexsFinal, cellsClean, None])
    structure = STRUCT([T(3)(0.1), structure])
    structure = OFFSET([.1, .1, .1])(structure)
    roof = STRUCT([COLOR([255 / 255., 204 / 255., 153 / 255., 1])(structure),
                   COLOR([229 / 255., 229 / 255., 229 / 255., 1])(skeleton1)])
    VIEW(roof)

ggpl_roof_builder(draw_roof())
```
